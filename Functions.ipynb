# create a module "comparison_operations"
 . it has 3 functions

 . 'greater_than' takes two parameters and return

   . 'greater_than(3,10)' returns '10 is greater than 3'

 . less than takes two parameters and output similar to above

 . equals return whether x equals y or x not equals y

```python
%%writefile comparision_operators.py
def greater_than(x,y):
    """ This method is used to know gratest number from two integer numbers"""
    if x>y:
        return f"{x} is greater than {y}"
    else:
        return f"{y} is greater than {x}"

def less_than(x,y):
    """ This method is used to know smallest number from two integer numbers"""
    if x>y:
        return f"{x} is less than {y}"
    else:
        return f"{y} is less than {x}"

def equals(x,y):
    """ This method is used to know  two integer numbers are equal or not"""
    if x==y:
        return f"{x} is equal to {y}"
    else:
        return f"{y} not equal to {x}"
```

# Filter Function
filter() function applies the predicate (condition) on sequence. If the predicate returns True, the value passed the test and is added to a filter object. Otherwise, the value is dropped from consideration.

```python
num=[34,90,41,79,52,16,23]
even=filter(lambda x:x%2==0,num)
odd=filter(lambda x:x%2!=0,num)
print(list(even))
print(list(odd))
```

# Reduce function
 . reduce() function from functools module is used to apply a function repeatedly to accumulate a single data value. A summation is a good exmpample of this process. The first value is added to the second value, then that sum is added to third value, and so on, until the sum of all the values is produced.

```python
from functools import reduce
```

```python
num=[45,71,90,32,39,42]
total=reduce(lambda x,y:x+y ,num)
total
```

# Recursion

```python
def factorial(n):
    if n==0 or n==1:
        return 1
    else:
        return n*factorial(n-1)
```

```python
factorial(5)
```

# *args and *kwargs
 . *args-when youhave no idea about how many arguments you will be receving as an input for the function.

 . **kwarga....how many keyword arguments you will be.....
 

```python
def exp(e,v,s):
    expression=e*v+s
    return expression
```

```python
def exp(e,v,s,*args):
    expression=e*v+s
    total=args[-1]*args[3]+args[2]
    return total
```

```python
exp(30,67,59,20,-1,20,37,45,89,93)
```

```python
def average(*args):
    return sum(args)/len(args)
```

```python
average(25,24,28,29,31,21,34,45,20,34)
```

```python
def exp(e,v,s,*args,**kwargs):
    expression=e*v+s
    total=args[-1]*args[3]+args[2]
    for k,v in kwargs.items():
        print(f"key:{k},value:{v}")
    return total
```

# prime number 

```python
from math import sqrt,floor
def is_prime(n):
    for i in range(2,floor(sqrt(n)+1)):
        if n%i==0:
            print(f"{n} is divisible by {i}. {n} is not a prime number")
            break
    else:
        print(f"{n} is a prime number")
        print("Else executed")
    print("Function completed")
```

```python
is_prime(23)
```
