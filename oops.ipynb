# Multilevel Inheritence:
. This occurs when class is derived from a class that itself is derived from a class that itself is derived from another class.For example, `LuxuryCar` inherits from `ElectricCar` which in-turn inherits from `Car` .This creates chain of hierarchy of in heritence allowing for more specialized behaviour at each level.

```python
class Car:
    def __init__(self,brand,model,year):
        self.brand=brand
        self.model=model
        self.year=year
    def details(self):
        print(f"{self.brand} car is {self.model} model and interduced in {self.year}")  
```

```python
car1=Car("BMW","Model S5",2021)
```

```python
class ElectricCar(Car):
    def __init__(self,brand,model,year,battery_capacity):
        super().__init__(brand,model,year)
        self.battery_capacity=battery_capacity
    def refuel(self):
         print(f"{self.brand} car is {self.model} is charging to {self.battery_capacity} KWh")
        
```

```python
ec1=ElectricCar("Tesla","Model s",2023,75)
```

```python
ec1.details()
```

```python
ec1.refuel()
```

```python
class LuxuryCar(ElectricCar):
     def __init__(self,brand,model,year,battery_capacity,luxury_features):
        super().__init__(brand,model,year,battery_capacity)
        self.luxury_features=luxury_features
     def luxury_features_available(self):
        print(f"Available luxury features : {self.luxury_features}")
    
```

```python
class LandVehicle:
    def __init__(self,land_speed):
        self.land_speed=land_speed
    def speed(self):
        print(f"Speed of the Vehicle : {self.land_speed}")
```

```python
class WaterVehicle:
    def __init__(self,water_speed):
        self.water_speed=water_speed
    def speed(self):
        print(f"Speed of the Vehicle : {self.water_speed}")
```

```python
class AmphibiousVehicle(LandVehicle, WaterVehicle):
    def __init__(self,land_speed,water_speed):
       LandVehicle.__init__(self,land_speed)
       WaterVehicle.__init__(self,water_speed)
```

```python
AmphibiousVehicle.__mro__
```

# Create multiple inheritence with vehicle as priority for methods

```python
class AmphibiousWaterVehicle(WaterVehicle,LandVehicle):
    def __init__(self,land_speed,water_speed):
       LandVehicle.__init__(self,land_speed)
       WaterVehicle.__init__(self,water_speed)
```

```python
amp1=AmphibiousVehicle(40,25)    #this is object for class with landvehicle class as priority
```

```python
amw1=AmphibiousWaterVehicle(50,30)   #this is object for class with watervehicle class as priority.
```

```python
amp1.speed()
```

```python
amw1.speed()
```

```python
class Counter:
    def __init__(self,max_count):
        self.__count=0    #private variable not accessible outside class
        self.__max_count=max_count
    def __auto_increment(self):   #private method not accessible outside class
        while self.__count<self.__max_count:
            self.__count+=1
            print(f"Current count :{self.__count}")
    def start_counting(self):
        self.__auto_increment()
        print("Maximum count reached")
```

```python
counter_1=Counter(20)
```

# Abstraction
` hidding unneccecary info

` showing only neccesary information

```python
from abc import ABC, abstractmethod
```

```python
print(dir(ABC))
```

```python
class MobilePhone(ABC): #MobilePhone is inherited from ABC(Abstract Base Class)
    @abstractmethod
    def call(self,number):
        """
        Make a phone call to a given number
        """
        pass  #Do nothing
    @abstractmethod
    def message(self,number,message):
        """
        Send a message to given number
        """
        pass
    @abstractmethod
    def turn_on(self):
        """
        Turn on the mobile phone
        """
        pass
    @abstractmethod
    def turn_off(self):
        """
        Turn off the mobile phone
        """
        pass
        
```

# if any method is declared as @abstractmethod in super class it should be implemented in sub class/child class

```python
class MySmartPhone(MobilePhone):
    def call(self,number):
        print(f"calling {number}....")
    def message(self,number,msg):
        print(f"sending a message to {number}  :{msg}")
    def turn_on(self):
        print("Mobile is turning on ...")
    def turn_off(self):
        print("Mobile is turning off...")       
```

```python
android=MySmartPhone()
```

```python
android.message(23,"Hi")
```
